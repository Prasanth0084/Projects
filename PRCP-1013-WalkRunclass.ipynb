```python
# Feature Selection
fs= data[['acceleration_y', 'acceleration_z', 'gyro_x', 'gyro_y']]
fs
```

```python
#standardize features by applying scaling
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
fd=Scaler.fit_transform(fs)
```

```python
fd
```

```python
res=pd.DataFrame(fd,columns=['acceleration_y', 'acceleration_z', 'gyro_x', 'gyro_y'])
res
```

```python
final_df=pd.concat([res,data['activity']],axis=1)
final_df
```

```python
x=final_df.iloc[:,:-1]
x
```

```python
y=final_df['activity']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Model: Random Forest
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(x_train,y_train)
```

```python
rf_pred=rf_model.predict(x_test)
rf_pred
```

```python
from sklearn.metrics import*
accuracy_score(y_test,rf_pred)
```

```python
print(classification_report(y_test,rf_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,rf_pred),annot=True)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(x_train,y_train)
```

```python
svc_pred=svc_model.predict(x_test)
svc_pred
```

```python
accuracy_score(y_test,svc_pred)
```

```python
print(classification_report(y_test,svc_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,svc_pred),annot=True)
```

```python
# Define the Deep Learning Model
import tensorflow as tf
from tensorflow import keras
model = keras.Sequential([
    keras.layers.Dense(64, activation='relu', input_shape=(x_train.shape[1],)),
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid') 
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(x_train, y_train, epochs=10, batch_size=32, validation_data=(x_test, y_test), verbose=1)
```

```python
# Evaluate the model
nn_pred = (model.predict(x_test) > 0.5).astype("int32")
```

```python
accuracy_score(y_test,nn_pred)
```

```python
print(classification_report(y_test,nn_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,nn_pred),annot=True)
```
